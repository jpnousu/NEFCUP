```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "/scratch/project_2000908/nousu/mNFI_to_NEP/FINAL_LAYERS/upscaling_layers.nc"
data = xr.open_dataset(file)
```

```python
data
```

```python
fig_file = r'/scratch/project_2000908/nousu/mNFI_to_NEP/figs.png'
# Get all data variables
all_vars = list(data.data_vars)

# Determine grid size
n_vars = len(all_vars)
n_cols = 4  # You can change this to e.g., 5 or 6 if you want a wider layout
n_rows = int(np.ceil(n_vars / n_cols))

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(4 * n_cols, 5 * n_rows))
axes = axes.flatten()

# Plot each variable
for i, var in enumerate(all_vars):
    data[var].plot(ax=axes[i], cmap="viridis", cbar_kwargs={'shrink': 0.7})
    axes[i].set_title(var)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("")

# Hide unused axes
for j in range(n_vars, len(axes)):
    axes[j].axis("off")
    
plt.savefig(fig_file, dpi=300, bbox_inches='tight')
```
